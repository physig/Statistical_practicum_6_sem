## Problem 1

Make a tuple containing natural numbers, the square of which is a multiple of 3, 4, but not a multiple of 8 and not exceeding 12345.

````python
limit = int(12345 ** 0.5)
result_list = []
for num in range(1, limit + 1):
    square = num * num
    if square % 3 == 0 and square % 4 == 0 and square % 8 != 0:
        result_list.append(num)
final_tuple = tuple(result_list)
print(final_tuple)
````

## Problem 2


Write a function that takes a two-dimensional array and a string as input and returns an array rotated 90 degrees counterclockwise if the string 'left' was passed, and clockwise if the string 'right' was passed.

Example for input: $\begin{bmatrix} 1 & 2 & 3 \\ 4 & 5 & 6 \\ 7 & 8 & 9 \end{bmatrix}$.\
If the string 'left' is passed, the function should return $\begin{bmatrix} 3 & 6 & 9 \\ 2 & 5 & 8 \\ 1 & 4 & 7 \end{bmatrix}$, and if the string 'right' is passed, the function should return $\begin{bmatrix} 7 & 4 & 1 \\ 8 & 5 & 2 \\ 9 & 6 & 3 \end{bmatrix}$.

Поворот матрицы с помощью прямого пересчета индексов

````python
def rotate(arr, direction):
    n = len(arr)          # количество строк
    m = len(arr[0])       # количество столбцов
    
    # Создаём матрицу для результата размером m x n
    result = [[0] * n for _ in range(m)]
    
    if direction == 'right':
        # При повороте вправо:
        # элемент (i, j) переходит в (j, n-1-i)
        for i in range(n):
            for j in range(m):
                result[j][n - 1 - i] = arr[i][j]
                
    elif direction == 'left':
        # При повороте влево:
        # элемент (i, j) переходит в (m-1-j, i)
        for i in range(n):
            for j in range(m):
                result[m - 1 - j][i] = arr[i][j]
    else:
        return "Ошибка: неверное направление"
    
    return result

# Проверка на неквадратной матрице
matrix_rect = [[1, 2, 3, 4],
               [5, 6, 7, 8]]

print("Исходная матрица 2x4:")
for row in matrix_rect:
    print(row)

print("\nПоворот вправо (4x2):")
rotated_right = rotate(matrix_rect, 'right')
for row in rotated_right:
    print(row)

print("\nПоворот влево (4x2):")
rotated_left = rotate(matrix_rect, 'left')
for row in rotated_left:
    print(row)
````

## Problem 3

Write a function that takes a string as input and returns a dictionary containing the number of occurrences of each character in the string.

Example for the string 'hello, world!': {'h': 1, 'e': 1, 'l': 3, 'o': 2, ',': 1, ' ': 1, 'w': 1, 'r': 1, 'd': 1, '!': 1}.

````python
def char_counter(string):
    result_dict = {}
    for ch in string:
        if ch in result_dict:
            result_dict[ch] += 1
        else:
            result_dict[ch] = 1
    return result_dict

char_counter("hello world!")
````

## Problem 4

### Implementing a Library Management System

#### Description

You are required to design and implement a system for managing books and users in a library. The system should allow for the management of books (adding, deleting, searching by various criteria) and users (registration, deletion, searching), as well as tracking the history of interactions between them (issuing and returning books).

#### Tasks

1. **`Book` Class**:
   - Attributes: title, author, year of publication, ISBN, number of copies.
   - Methods: constructor, methods to get information about the book, method to change the number of copies (when issuing and returning books).

2. **`User` Class**:
   - Attributes: user name, library card number, list of borrowed books.
   - Methods: constructor, methods for user registration, methods for adding and removing books from the borrowed list.

3. **`Library` Class**:
   - Attributes: list of books, list of users, transaction history (who, when, which book was borrowed and returned).
   - Methods: constructor, methods for adding and deleting books and users, methods for issuing and returning books, searching for books and users by various criteria, method to display the transaction history.

#### Assignment

1. Implement the `Book`, `User`, and `Library` classes with the specified attributes and methods.
2. Create several books and users, and add them to the library system.
3. Implement scenarios for issuing books to users and their return.
4. Display the transaction history to show how books were issued and returned.


````python
import datetime

class Book:
    def __init__(self, title, author, year, isbn, quantity):
        self.title = title
        self.author = author
        self.year = year
        self.isbn = isbn
        self.quantity = quantity
        print(f"Книга '{title}' добавлена")
    
    def get_info(self):
        return f"Название: {self.title}, Автор: {self.author}, Год: {self.year}, ISBN: {self.isbn}, Экземпляров: {self.quantity}"
    
    def change_quantity(self, action):
        if action == 'issue':
            self.quantity -= 1
        elif action == 'return':
            self.quantity += 1

class User:
    def __init__(self, name, card_number):
        self.name = name
        self.card_number = card_number
        self.borrowed_books = []
        print(f"Пользователь '{name}' добавлен")
    
    def register(self):
        return f"Пользователь {self.name} зарегистрирован, номер карты: {self.card_number}"
    
    def add_borrowed(self, book):
        self.borrowed_books.append(book)
    
    def remove_borrowed(self, book):
        if book in self.borrowed_books:
            self.borrowed_books.remove(book)
    
    def display_info(self):
        borrowed_titles = [book.title for book in self.borrowed_books]
        return f"Имя: {self.name}, Номер карты: {self.card_number}, Взятые книги: {', '.join(borrowed_titles) if borrowed_titles else 'нет'}"

class Library:
    def __init__(self):
        self.books = []
        self.users = []
        self.history = []
        print("Библиотека создана")
    
    def add_book(self, book):
        self.books.append(book)
    
    def remove_book(self, book):
        if book in self.books:
            self.books.remove(book)
    
    def add_user(self, user):
        self.users.append(user)
    
    def remove_user(self, user):
        if user in self.users:
            self.users.remove(user)
    
    def issue_book(self, user, book):
        if book.quantity > 0 and user in self.users:
            user.add_borrowed(book)
            book.change_quantity('issue')
            self.history.append(f"пользователь '{user.name}' взял книгу '{book.title}' в {datetime.datetime.now()}")
            print("книга успешно выдана")
        else:
            print(f"ошибка: книга '{book.title}' недоступна или пользователь '{user.name}' не найден")
    
    def return_book(self, user, book):
        if book in user.borrowed_books and user in self.users:
            user.remove_borrowed(book)
            book.change_quantity('return')
            self.history.append(f"пользователь '{user.name}' вернул книгу '{book.title}' в {datetime.datetime.now()}")
            print("книга успешно возвращена")
        else:
            print(f"ошибка: книга '{book.title}' не была взята или пользователь '{user.name}' не найден")
    
    def show_history(self):
        print("история операций:")
        for entry in self.history:
            print(entry)
    
    def search_books_by_title(self, query):
        results = [b for b in self.books if query.lower() in b.title.lower()]
        print(f"по запросу '{query}' найдены книги:")
        for r in results:
            print(r.get_info())
    
    def search_books_by_author(self, query):
        results = [b for b in self.books if query.lower() in b.author.lower()]
        print(f"по запросу '{query}' найдены книги:")
        for r in results:
            print(r.get_info())
    
    def search_users_by_name(self, query):
        results = [u for u in self.users if query.lower() in u.name.lower()]
        print(f"по запросу '{query}' найдены пользователи:")
        for r in results:
            print(r.display_info())
    
    def search_users_by_card(self, card):
        results = [u for u in self.users if card == u.card_number]
        print(f"по запросу '{card}' найдены пользователи:")
        for r in results:
            print(r.display_info())
    
    def search_users_by_borrowed_book(self, title):
        results = [u for u in self.users for b in u.borrowed_books if title.lower() in b.title.lower()]
        print(f"по запросу '{title}' найдены пользователи, взявшие эту книгу:")
        for r in results:
            print(r.display_info())

# Пример использования - новые книги и пользователи
library = Library()

book1 = Book('451 градус по Фаренгейту', 'Рэй Брэдбери', 1953, '978-5-17-118759-7', 8)
book2 = Book('Сто лет одиночества', 'Габриэль Гарсиа Маркес', 1967, '978-5-389-08472-8', 6)
book3 = Book('Улисс', 'Джеймс Джойс', 1922, '978-5-699-63578-4', 3)
book4 = Book('Преступление и наказание', 'Фёдор Достоевский', 1866, '978-5-17-082456-2', 10)

user1 = User('Екатерина Смирнова', '202401')
user2 = User('Иван Козлов', '202402')
user3 = User('Ольга Новикова', '202403')
user4 = User('Максим Воробьёв', '202404')

for b in [book1, book2, book3, book4]:
    library.add_book(b)
for u in [user1, user2, user3, user4]:
    library.add_user(u)

print("\n")

library.issue_book(user1, book2)
library.issue_book(user2, book1)
library.issue_book(user3, book4)
library.issue_book(user4, book3)
library.return_book(user1, book2)
library.return_book(user2, book1)
library.issue_book(user2, book4)
library.issue_book(user1, book3)

print("\n")
library.show_history()

print("\n")
library.search_books_by_title('одиночества')
library.search_books_by_author('Достоевский')
library.search_users_by_name('ова')
library.search_users_by_card('202403')
library.search_users_by_borrowed_book('Преступление')
````

## Problem 5*

Explain why list `b` changes after the execution of the following code:

```python
a = [1, 2, 3]
b = a
a[0] = 4
print(b)
```

````python
a = [1, 2, 3]
b = a
a[0] = 4
print(b)
````

> Write your answer in markdown cell after:

Переменная `b` становится ссылкой на тот же объект списка, что и `a`. В Python списки относятся к изменяемым типам данных (mutable). Когда мы меняем элемент через `a[0] = 4`, мы изменяем сам объект в памяти, на который указывают обе переменные. Поэтому при выводе `b` мы видим изменённый список. Чтобы создать независимую копию, нужно использовать `b = a.copy()` или `b = a[:]`.

## Problem 6*

Let
$$A = \sum_{i=1}^{10000} \frac{1}{i^2},\quad B=\sum_{i=10000}^{1} \frac{1}{i^2}.$$
Calculate the values of $A$ and $B$ and compare them. What do you observe? Explain why this happens. What is the best way to calculate the value of $\sum\limits_{i=1}^{10000} \dfrac{1}{i^2}$?

````python
sum_forward = 0.0
for i in range(1, 10001):
    sum_forward += 1.0 / (i * i)
print(sum_forward)

sum_backward = 0.0
for i in range(10000, 0, -1):
    sum_backward += 1.0 / (i * i)
print(sum_backward)
````

Сумма, вычисленная от больших членов к малым, оказывается немного точнее. Это связано с тем, что при сложении чисел с плавающей точкой сначала нужно складывать маленькие числа, чтобы они не "терялись" на фоне больших. Поэтому оптимальный способ — суммировать в обратном порядке: от самого маленького слагаемого к самому большому.
