## Задача 1

Пусть $f(x) = 2\sin x - \cos{2x}$. 

Написать функцию `beauty_derivative_plot`, которая принимает точку $x$ и строит графики функции $f(x)$, её производной $f'(x)$ и касательной в точке $x$.

```python
import numpy as np
import matplotlib.pyplot as plt

def beauty_derivative_plot(x0):
    """
       
    Parameters:
    -----------
    x0 : float
        Точка, в которой строится касательная
    """
    # Определяем функцию и её производную
    def f(x):
        return 2 * np.sin(x) - np.cos(2 * x)

    def df(x):
        # Производная: d/dx[2sin(x)] = 2cos(x)
        # d/dx[-cos(2x)] = -(-sin(2x))*2 = 2sin(2x)
        return 2 * np.cos(x) + 2 * np.sin(2 * x)

    # Создаем массив x для построения графиков
    x = np.linspace(-2 * np.pi, 2 * np.pi, 500)

    # Вычисляем значения функции и её производной
    y = f(x)
    dy = df(x)

    # Вычисляем касательную в точке x0
    y0 = f(x0)
    slope = df(x0)

    # Строим касательную на интервале [x0 - 2, x0 + 2]
    x_tangent = np.linspace(x0 - 2, x0 + 2, 100)
    y_tangent = slope * (x_tangent - x0) + y0

    # Строим график
    plt.figure(figsize=(12, 8))

    # Основная функция
    plt.plot(x, y, 'b-', linewidth=2.5, label='$f(x) = 2\sin(x) - \cos(2x)$')

    # Производная
    plt.plot(x, dy, 'g--', linewidth=2, label="$f'(x) = 2\cos(x) + 2\sin(2x)$")

    # Касательная
    plt.plot(x_tangent, y_tangent, 'r-', linewidth=2, label=f'Касательная в x = {x0:.2f}')

    # Отмечаем точку касания
    plt.plot(x0, y0, 'ro', markersize=10, markeredgecolor='black', markeredgewidth=1.5)

    # Добавляем вертикальную линию в точке касания
    plt.axvline(x=x0, color='gray', linestyle=':', alpha=0.5)

    # Настройка графика
    plt.title(f'График функции $f(x)$ и её касательной в точке x = {x0:.2f}', fontsize=14)
    plt.xlabel('x', fontsize=12)
    plt.ylabel('y', fontsize=12)
    plt.grid(True, alpha=0.3)
    plt.legend(fontsize=11)

    # Устанавливаем пределы осей
    plt.xlim(-2 * np.pi, 2 * np.pi)
    plt.ylim(-5, 5)

    plt.tight_layout()
    plt.show()


# Примеры вызова функции
print("\n" + "="*50)
print("Пример 1: x = -3")
print("="*50)
beauty_derivative_plot(-3)

print("\n" + "="*50)
print("Пример 2: x = -2")
print("="*50)
beauty_derivative_plot(-2)

print("\n" + "="*50)
print("Пример 3: x = 1.7")
print("="*50)
beauty_derivative_plot(1.7)
```

## Задача 2

Анализ данных о популяциях зайцев, рысей и моркови в Северной Канаде.

**Важно:** Файл `populations.txt` должен находиться в той же папке, что и этот ноутбук!

```python
import numpy as np
import matplotlib.pyplot as plt
import os

# Проверяем наличие файла
if not os.path.exists('populations.txt'):
    print("ОШИБКА: Файл 'populations.txt' не найден!")
    print("Убедитесь, что файл находится в той же папке, что и этот ноутбук.")
else:
    # Загружаем данные из локального файла
    data = np.loadtxt('populations.txt')
    years = data[:, 0].astype(int)
    hares = data[:, 1]
    lynxes = data[:, 2]
    carrots = data[:, 3]

    print("Данные успешно загружены из populations.txt!")
    print(f"Годы: {years}")
    print(f"Зайцы: {hares}")
    print(f"Рыси: {lynxes}")
    print(f"Морковь: {carrots}")
```

### 1. График популяций всех видов

```python
if os.path.exists('populations.txt'):
    plt.figure(figsize=(14, 7))
    plt.plot(years, hares, 'b-', marker='o', linewidth=2, markersize=8, label='Зайцы')
    plt.plot(years, lynxes, 'r-', marker='s', linewidth=2, markersize=8, label='Рыси')
    plt.plot(years, carrots, 'g-', marker='^', linewidth=2, markersize=8, label='Морковь')

    plt.xlabel('Год', fontsize=12)
    plt.ylabel('Популяция', fontsize=12)
    plt.title('Динамика популяций зайцев, рысей и моркови', fontsize=14)
    plt.grid(True, alpha=0.3)
    plt.legend(fontsize=11)
    plt.tight_layout()
    plt.show()
```

### 2. Годы максимальной популяции каждого вида

```python
if os.path.exists('populations.txt'):
    # Находим индексы максимальных значений
    max_hare_idx = np.argmax(hares)
    max_lynx_idx = np.argmax(lynxes)
    max_carrot_idx = np.argmax(carrots)

    print(f"Максимальная популяция зайцев: {hares[max_hare_idx]} в {years[max_hare_idx]} году")
    print(f"Максимальная популяция рысей: {lynxes[max_lynx_idx]} в {years[max_lynx_idx]} году")
    print(f"Максимальная популяция моркови: {carrots[max_carrot_idx]} в {years[max_carrot_idx]} году")

    # Строим график с отмеченными максимумами
    plt.figure(figsize=(14, 7))
    plt.plot(years, hares, 'b-', marker='o', linewidth=2, markersize=6, label='Зайцы')
    plt.plot(years, lynxes, 'r-', marker='s', linewidth=2, markersize=6, label='Рыси')
    plt.plot(years, carrots, 'g-', marker='^', linewidth=2, markersize=6, label='Морковь')

    # Отмечаем максимальные точки
    plt.plot(years[max_hare_idx], hares[max_hare_idx], 'bo', markersize=12, markeredgecolor='black', markeredgewidth=2)
    plt.plot(years[max_lynx_idx], lynxes[max_lynx_idx], 'ro', markersize=12, markeredgecolor='black', markeredgewidth=2)
    plt.plot(years[max_carrot_idx], carrots[max_carrot_idx], 'go', markersize=12, markeredgecolor='black', markeredgewidth=2)

    # Добавляем текстовые метки
    plt.text(years[max_hare_idx] + 0.5, hares[max_hare_idx] + 50, f'{hares[max_hare_idx]:.0f}', fontsize=10)
    plt.text(years[max_lynx_idx] + 0.5, lynxes[max_lynx_idx] + 50, f'{lynxes[max_lynx_idx]:.0f}', fontsize=10)
    plt.text(years[max_carrot_idx] + 0.5, carrots[max_carrot_idx] + 50, f'{carrots[max_carrot_idx]:.0f}', fontsize=10)

    plt.xlabel('Год', fontsize=12)
    plt.ylabel('Популяция', fontsize=12)
    plt.title('Динамика популяций с отмеченными максимумами', fontsize=14)
    plt.grid(True, alpha=0.3)
    plt.legend(fontsize=11)
    plt.tight_layout()
    plt.show()
```

### 3. Stacked bar chart: соотношение зайцев и рысей

```python
if os.path.exists('populations.txt'):
    # Вычисляем общую популяцию зайцев и рысей
    total = hares + lynxes

    # Вычисляем долю каждого вида (в процентах)
    hare_ratio = hares / total * 100
    lynx_ratio = lynxes / total * 100

    plt.figure(figsize=(14, 7))
    plt.bar(years, hare_ratio, label='Зайцы', color='blue', alpha=0.8)
    plt.bar(years, lynx_ratio, bottom=hare_ratio, label='Рыси', color='red', alpha=0.8)

    plt.xlabel('Год', fontsize=12)
    plt.ylabel('Процент от общей популяции', fontsize=12)
    plt.title('Соотношение популяций зайцев и рысей (в процентах)', fontsize=14)
    plt.legend(fontsize=11)
    plt.grid(True, alpha=0.3, axis='y')
    plt.tight_layout()
    plt.show()
```

### 4. Гистограммы популяций со средними значениями

```python
if os.path.exists('populations.txt'):
    fig, axes = plt.subplots(1, 3, figsize=(16, 5))

    # Статистика
    hare_mean, hare_std = np.mean(hares), np.std(hares)
    lynx_mean, lynx_std = np.mean(lynxes), np.std(lynxes)
    carrot_mean, carrot_std = np.mean(carrots), np.std(carrots)

    # Гистограмма для зайцев
    axes[0].hist(hares, bins=8, color='blue', alpha=0.7, edgecolor='black', linewidth=1.2)
    axes[0].axvline(hare_mean, color='red', linestyle='--', linewidth=2, label=f'Среднее: {hare_mean:.1f}')
    axes[0].set_title(f'Зайцы\nСр.={hare_mean:.1f}, Стд.={hare_std:.1f}', fontsize=12)
    axes[0].set_xlabel('Популяция', fontsize=11)
    axes[0].set_ylabel('Частота', fontsize=11)
    axes[0].legend()
    axes[0].grid(True, alpha=0.3)

    # Гистограмма для рысей
    axes[1].hist(lynxes, bins=8, color='red', alpha=0.7, edgecolor='black', linewidth=1.2)
    axes[1].axvline(lynx_mean, color='blue', linestyle='--', linewidth=2, label=f'Среднее: {lynx_mean:.1f}')
    axes[1].set_title(f'Рыси\nСр.={lynx_mean:.1f}, Стд.={lynx_std:.1f}', fontsize=12)
    axes[1].set_xlabel('Популяция', fontsize=11)
    axes[1].set_ylabel('Частота', fontsize=11)
    axes[1].legend()
    axes[1].grid(True, alpha=0.3)

    # Гистограмма для моркови
    axes[2].hist(carrots, bins=8, color='green', alpha=0.7, edgecolor='black', linewidth=1.2)
    axes[2].axvline(carrot_mean, color='red', linestyle='--', linewidth=2, label=f'Среднее: {carrot_mean:.1f}')
    axes[2].set_title(f'Морковь\nСр.={carrot_mean:.1f}, Стд.={carrot_std:.1f}', fontsize=12)
    axes[2].set_xlabel('Популяция', fontsize=11)
    axes[2].set_ylabel('Частота', fontsize=11)
    axes[2].legend()
    axes[2].grid(True, alpha=0.3)

    plt.tight_layout()
    plt.show()

    print(f"\nСтатистика:")
    print(f"Зайцы: среднее = {hare_mean:.1f}, стандартное отклонение = {hare_std:.1f}")
    print(f"Рыси: среднее = {lynx_mean:.1f}, стандартное отклонение = {lynx_std:.1f}")
    print(f"Морковь: среднее = {carrot_mean:.1f}, стандартное отклонение = {carrot_std:.1f}")
```

### 5. Сравнение популяций зайцев и рысей

```python
if os.path.exists('populations.txt'):
    # Сравниваем популяции зайцев и рысей
    comparison = np.where(hares > lynxes, 'H', 'L')
    print(f"Результат сравнения (H - зайцы больше, L - рыси больше): {list(comparison)}")

    # количество лет, когда зайцев было больше
    hare_wins = np.sum(hares > lynxes)
    lynx_wins = np.sum(lynxes > hares)
    tie = np.sum(hares == lynxes)

    print(f"\nКоличество лет, когда зайцев больше: {hare_wins}")
    print(f"Количество лет, когда рысей больше: {lynx_wins}")
    print(f"Количество лет с равной популяцией: {tie}")

   
    plt.figure(figsize=(8, 8))
    plt.pie([hare_wins, lynx_wins], labels=['Зайцы больше', 'Рыси больше'], 
            colors=['blue', 'red'], autopct='%1.1f%%', startangle=90, explode=(0.05, 0))
    plt.title('Соотношение лет с большей популяцией зайцев и рысей', fontsize=14)
    plt.tight_layout()
    plt.show()
```

### 6. Изменение популяций и корреляция

```python
if os.path.exists('populations.txt'):
   
    hare_change = np.diff(hares)
    lynx_change = np.diff(lynxes)


    plt.figure(figsize=(14, 7))
    plt.plot(years[1:], hare_change, 'b-', marker='o', linewidth=2, markersize=8, label='Изменение зайцев')
    plt.plot(years[1:], lynx_change, 'r-', marker='s', linewidth=2, markersize=8, label='Изменение рысей')

    plt.xlabel('Год', fontsize=12)
    plt.ylabel('Изменение популяции', fontsize=12)
    plt.title('Изменение популяций зайцев и рысей по годам', fontsize=14)
    plt.grid(True, alpha=0.3)
    plt.legend(fontsize=11)
    plt.tight_layout()
    plt.show()

    # Вычисляем коэффициент корреляции
    correlation = np.corrcoef(hare_change, lynx_change)[0, 1]
    print(f"Коэффициент корреляции между изменениями популяций зайцев и рысей: {correlation:.3f}")

    # Интерпретация
    if correlation > 0.7:
        print("Сильная положительная корреляция")
    elif correlation > 0.3:
        print("Слабая положительная корреляция")
    elif correlation > -0.3:
        print("Корреляция отсутствует")
    elif correlation > -0.7:
        print("Слабая отрицательная корреляция")
    else:
        print("Сильная отрицательная корреляция")
```

### 7. Scatter plot: зайцы vs рыси

```python
if os.path.exists('populations.txt'):
    plt.figure(figsize=(12, 10))
    scatter = plt.scatter(hares, lynxes, c=years, cmap='viridis', s=150, alpha=0.7, edgecolors='black', linewidth=1.5)

  
    cbar = plt.colorbar(scatter)
    cbar.set_label('Год', fontsize=12)

    # y = x для сравнения
    max_val = max(np.max(hares), np.max(lynxes))
    min_val = min(np.min(hares), np.min(lynxes))
    plt.plot([min_val, max_val], [min_val, max_val], 'k--', alpha=0.5, linewidth=2, label='y = x')

   
    for i, year in enumerate(years):
        plt.annotate(str(int(year)), (hares[i], lynxes[i]), 
                    fontsize=10, alpha=0.8, xytext=(5, 5), textcoords='offset points')

    plt.xlabel('Популяция зайцев', fontsize=13)
    plt.ylabel('Популяция рысей', fontsize=13)
    plt.title('Зависимость популяции рысей от популяции зайцев', fontsize=14)
    plt.grid(True, alpha=0.3)
    plt.legend(fontsize=11)
    plt.tight_layout()
    plt.show()
```

### 8. Оценка популяции зайцев в 1920 году

```python
if os.path.exists('populations.txt'):
    print(f"Диапазон лет в данных: {years[0]} - {years[-1]}")
    print(f"Данные за 1920 год отсутствуют")

    
    coeffs = np.polyfit(years, hares, 2)
    poly = np.poly1d(coeffs)

    # Прогнозируем популяцию на 1920 год
    year_to_predict = 1920
    predicted = poly(year_to_predict)

    print(f"\nПрогнозируемая популяция зайцев в {year_to_predict} году: {predicted:.1f}")


    plt.figure(figsize=(14, 7))
    plt.plot(years, hares, 'bo-', linewidth=2, markersize=8, label='Фактические данные')

    # Строим кривую регрессии
    x_smooth = np.linspace(years[0], year_to_predict, 100)
    y_smooth = poly(x_smooth)
    plt.plot(x_smooth, y_smooth, 'r--', linewidth=2, label='Полиномиальная регрессия (2-я степень)')

    # Отмечаем прогнозируемую точку
    plt.plot(year_to_predict, predicted, 'ro', markersize=12, markeredgecolor='black', 
             markeredgewidth=2, label=f'Прогноз на {year_to_predict}: {predicted:.0f}')

    plt.xlabel('Год', fontsize=12)
    plt.ylabel('Популяция зайцев', fontsize=12)
    plt.title('Прогнозирование популяции зайцев на 1920 год', fontsize=14)
    plt.grid(True, alpha=0.3)
    plt.legend(fontsize=11)
    plt.tight_layout()
    plt.show()

  # линейная регрессия по последним 5 годам
    recent_years = years[-5:]
    recent_hares = hares[-5:]
    coeffs_linear = np.polyfit(recent_years, recent_hares, 1)
    poly_linear = np.poly1d(coeffs_linear)
    predicted_linear = poly_linear(year_to_predict)

    print(f"Прогноз по линейной регрессии (последние 5 лет): {predicted_linear:.1f}")
    print(f"Разница между методами: {abs(predicted - predicted_linear):.1f}")
```
