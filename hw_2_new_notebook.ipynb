- Find the mean and std of the populations of each species. Plot the historgam of population for each species and show mean values with vertical line. Show each histogram in a separate plot. Arrange 3 plots in a row.

```python
import numpy as np
import matplotlib.pyplot as plt
from google.colab import files
uploaded = files.upload()
u = np.loadtxt('populations.txt')
v = u[:, 1]
w = u[:, 2]
x = u[:, 3]
fig, (y, z, aa) = plt.subplots(1, 3, figsize=(15, 4))
bb = np.mean(v)
cc = np.std(v)
y.hist(v, bins=8, color='b', alpha=0.7, edgecolor='black')
y.axvline(bb, color='red', linewidth=2, label=f'Mean: {bb:.0f}')
y.set_title(f'Hares\nstd: {cc:.0f}')
y.legend()
dd = np.mean(w)
ee = np.std(w)
z.hist(w, bins=8, color='r', alpha=0.7, edgecolor='black')
z.axvline(dd, color='red', linewidth=2, label=f'Mean: {dd:.0f}')
z.set_title(f'Lynxes\nstd: {ee:.0f}')
z.legend()
ff = np.mean(x)
gg = np.std(x)
aa.hist(x, bins=8, color='g', alpha=0.7, edgecolor='black')
aa.axvline(ff, color='red', linewidth=2, label=f'Mean: {ff:.0f}')
aa.set_title(f'Carrots\nstd: {gg:.0f}')
aa.legend()
plt.tight_layout()
plt.show()
```

- Find which species (hares or lynxes) has the largest population for each year. Print the result as [H, H, L, H, ...]. Plot a pie chart showing the ratio of "H" and "L" values.

```python
import numpy as np
import matplotlib.pyplot as plt
from google.colab import files
uploaded = files.upload()
hh = np.loadtxt('populations.txt')
ii = hh[:, 1]
jj = hh[:, 2]
kk = ii > jj
ll = np.where(kk, 'H', 'L')
print("Species with larger population each year:")
print(ll)
mm = np.sum(kk)
nn = len(ii) - mm
plt.figure(figsize=(6, 6))
plt.pie([mm, nn], labels=['Hares dominate', 'Lynxes dominate'], autopct='%1.1f%%', colors=['b', 'r'])
plt.title('Which species dominates each year')
plt.show()
```

- Plot the change in the hare population and in the lynx population for each year. Find the correlation coefficient between the both time series.

```python
import numpy as np
import matplotlib.pyplot as plt
from google.colab import files
uploaded = files.upload()
oo = np.loadtxt('populations.txt')
pp = oo[:, 0]
qq = oo[:, 1]
rr = oo[:, 2]
ss = np.diff(qq)
tt = np.diff(rr)
plt.figure(figsize=(10, 6))
plt.plot(pp[1:], ss, 'b-o', label='Hare population change')
plt.plot(pp[1:], tt, 'r-s', label='Lynx population change')
plt.xlabel('Year')
plt.ylabel('Population change')
plt.title('Yearly population changes')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
uu = np.corrcoef(qq, rr)[0, 1]
print(f"Correlation coefficient between hare and lynx populations: {uu:.3f}")
```

- Assume the population of hares in 1920 is unknown. Suggest a way to estimate this value. Compare an estimated value
with the true value and print a ratio of the error to the true value.

```python
import numpy as np
from google.colab import files
uploaded = files.upload()
aaa = np.loadtxt('populations.txt')
bbb = aaa[:, 0]
ccc = aaa[:, 1]
ddd = bbb[bbb != 1920]
eee = ccc[bbb != 1920]
fff = np.polyfit(ddd, eee, 1)
ggg = np.poly1d(fff)
hhh = ggg(1920)
iii = ccc[bbb == 1920][0]
jjj = abs(hhh - iii) / iii
print(f"True 1920 hare population: {iii}")
print(f"Estimated 1920 hare population: {hhh:.0f}")
print(f"Error ratio: {jjj:.4f}")
```
